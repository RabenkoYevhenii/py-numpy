Rewrite Python code to `numpy` code (Masks).

### Task 7.1

````python
# Python code
def unknown_signature(lst, mask, new_value):
    for i in range(len(lst)):
        if mask[i]:
            lst[i] = new_value

lst = [1, 2, 3, 4, 5]
mask = [True, False, True, False, True]
new_value = 0

%timeit unknown_signature(lst, mask, new_value)
lst
````

````python
def unknown_signature_np(lst, mask, new_value):
    lst = np.array(lst)
    lst[mask] = new_value
    return lst.tolist()

lst = [1, 2, 3, 4, 5]
mask = [True, False, True, False, True]
new_value = 0

%timeit unknown_signature(lst, mask, new_value)
lst
````

### Task 7.2

````python
# Python code
def unknown_signature(lst1, lst2):
    bitwise_complement = [~x for x in lst1]
    bitwise_and = [x & y for x, y in zip(lst1, lst2)]
    bitwise_or = [x | y for x, y in zip(lst1, lst2)]
    return bitwise_complement, bitwise_and, bitwise_or

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]

result_bitwise_complement, result_bitwise_and, result_bitwise_or = unknown_signature(list1, list2)
%timeit result_bitwise_complement, result_bitwise_and, result_bitwise_or
````

````python
def unknown_signature_np(lst1, lst2):   
    bitwise_complement = np.bitwise_not(lst1)
    bitwise_and = np.bitwise_and(lst1, lst2)
    bitwise_or = np.bitwise_or(lst1, lst2)
    return bitwise_complement, bitwise_and, bitwise_or

list1 = np.array([5, 2, 7, 4, 9])
list2 = np.array([3, 6, 1, 8, 10])

result_bitwise_complement, result_bitwise_and, result_bitwise_or = unknown_signature(list1, list2)
%timeit result_bitwise_complement, result_bitwise_and, result_bitwise_or
````

### Task 7.3 Optional

````python
# Python code
def unknown_signature(lst):
    result = []
    for num in lst:
        if num % 2 == 0 and (num % 3 == 0 or num > 10):
            modified_num = (num ^ 7) + 2
            result.append(modified_num)
    return result

my_list = [6, 9, 12, 14, 17, 20]
result_python_style = unknown_signature(my_list)
%timeit result_python_style
````

````python
def unknown_signature_np(lst):
    lst = np.array(lst)
    mask = np.logical_and(lst % 2 == 0, np.logical_or(lst % 3 == 0, lst > 10))
    modified_nums = (np.bitwise_xor(lst[mask], 7) + 2).tolist()
    return modified_nums

my_list = [6, 9, 12, 14, 17, 20]
result_numpy_style = unknown_signature_np(my_list)
%timeit result_numpy_style
````
